Transformations to take question and response sets, shuffle responses and save output in csv that is used by the front end to surface quiz to user.

```python
import pandas as pd
import random
```

```python
raw = pd.read_csv('hierarchical_and_semantic_outputs_with_policy_numbers.csv', index_col = 0)
df = raw[['question', 'email_response', 'hierarchy_bot_response', 'semantic_bot_response', 'all_policies']].copy()
```

```python
eval(df.all_policies.iloc[10]).sort()
```

```python
def get_response_dict(human, hierarchy, semantic):
    res_list = [human, hierarchy, semantic]
    res_tuple = []
    # save into list of tuples
    for i, source in enumerate(['human', 'hierarchy', 'semantic']):
        res_tuple.append([res_list[i], source])

    # shuffle 
    random.shuffle(res_tuple)

    # append text with A. B. C. 
    i = 1
    for res in res_tuple:
        res[0] = f'**Option {i}:** ' + res[0]
        i += 1

    # save as a dictionary
    return dict(res_tuple)
```

```python
def sort_policies(all_policies):
    all_policies = eval(all_policies)
    all_policies.sort()

    return all_policies
```

```python
df['response_dict'] = df.apply(lambda x: get_response_dict(x.email_response, x.hierarchy_bot_response, x.semantic_bot_response), axis=1)
df['policies_sorted'] = df.apply(lambda x: sort_policies(x.all_policies), axis=1)
df
```

```python
df.drop(['email_response', 'hierarchy_bot_response', 'semantic_bot_response', 'all_policies'],errors='ignore',axis=1, inplace=True)
```

```python
df.to_csv('responses_to_surface.csv')
```
